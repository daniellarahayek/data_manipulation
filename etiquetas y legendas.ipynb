```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Daniel Lara/Downloads/2008.csv", nrows = 100000)
data = np.unique(df.DayOfWeek,return_counts = True)
labs= ["lun","Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x=data[1],
       labels = labs,
       radius=1.5,
       colors = sns.color_palette("hls",7),
       explode=(0,0,0,0,0,0,.1),
       startangle = 90,
       autopct = "%.2f")
# plt.legend(loc="right", labels = labs)
plt.show()
```

```python
plt = sns.barplot(x=labs, y=data[1])
plt.set(xlabel = "dia de la semana", ylabel="numero de vuelos")
```
