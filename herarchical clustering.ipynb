```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Daniel Lara/Downloads/2008.csv",nrows = 1e4)
newdf = df[["DepDelay","AirTime"]].dropna()
```

```python
newdf = df[["ArrDelay","Distance","DepDelay","TaxiOut","AirTime"]].dropna()
```

```python
clstr = AgglomerativeClustering(n_clusters = 5)
clstr.fit(newdf)

# euclidean/ L1/L2/manhattan/cosine/precomputed
#affinity="L1",linkage="Complete"
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c = clstr.fit_predict(newdf))
plt.show()
```
