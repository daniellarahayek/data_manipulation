```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Daniel Lara/Downloads/2008.csv",nrows = 1e5)
newdf = df[["DepDelay","AirTime"]].dropna()
```

```python
newdf = df[["ArrDelay","Distance","DepDelay","TaxiOut","AirTime"]].dropna()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0).fit(newdf)
kmeans.labels_
```

```python
np.unique(kmeans.labels_,return_counts = True)
```

```python
plt.scatter(newdf["AirTime"],newdf["DepDelay"],c = kmeans.labels_)
plt.show()
```

```python
kmeans.cluster_centers_
```
